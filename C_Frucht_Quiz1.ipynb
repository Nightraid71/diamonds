```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/Nightraid71/diamonds/refs/heads/main/diamonds.csv'

diamonds = pd.read_csv(url)
print(diamonds.head())

#part1  # Gets the number of cases, attributes,and data type from the data
print("Number of cases:", diamonds.shape[0])
print("Number of attributes:", diamonds.shape[1])
print("Attribute & Data types:")
print(diamonds.dtypes)

#part 2 # Gets the statistics of the attributes with graphical visual
print("Summary Statistics:\n")
print(diamonds.describe(include='all'))

print("\nCarat Summary:\n", diamonds['carat'].describe())
plt.hist(diamonds['carat'], bins=30, color='skyblue', edgecolor='black')
plt.title('Carat Distribution (Matplotlib)')
plt.xlabel('Carat')
plt.ylabel('Frequency')
plt.show()
sns.histplot(diamonds['carat'], bins=30, kde=True, color='skyblue')
plt.title('Carat Distribution (Seaborn)')
plt.xlabel('Carat')
plt.ylabel('Frequency')
plt.show()
# Most diamonds weigh less than 1 carat, While the larger diamonds pull the average higher

print("\nCut Value Counts:\n", diamonds['cut'].value_counts())
diamonds['cut'].value_counts().plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Cut Distribution (Matplotlib)')
plt.xlabel('Cut')
plt.ylabel('Count')
plt.show()
sns.countplot(data=diamonds, x='cut', palette='Set2')
plt.title('Cut Distribution (Seaborn)')
plt.xlabel('Cut')
plt.ylabel('Count')
plt.show()
# Very few diamonds have a fair cut, with the most common being ideal, premium followed by very good

print("\nColor Value Counts:\n", diamonds['color'].value_counts())
diamonds['color'].value_counts().sort_index().plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Color Distribution (Matplotlib)')
plt.xlabel('Color Grade')
plt.ylabel('Count')
plt.show()
sns.countplot(data=diamonds, x='color', order=sorted(diamonds['color'].unique()))
plt.title('Color Distribution (Seaborn)')
plt.xlabel('Color Grade')
plt.ylabel('Count')
plt.show()
# Colors range from best to least desirable 

print("\nClarity Value Counts:\n", diamonds['clarity'].value_counts())
diamonds['clarity'].value_counts().sort_index().plot(kind='bar', color='red', edgecolor='black')
plt.title('Clarity Distribution (Matplotlib)')
plt.xlabel('Clarity Grade')
plt.ylabel('Count')
plt.show()
sns.countplot(data=diamonds, x='clarity', order=sorted(diamonds['clarity'].unique()))
plt.title('Clarity Distribution (Seaborn)')
plt.xlabel('Clarity Grade')
plt.ylabel('Count')
plt.show()
# Grades SI1 & VS2 are the most common with few diamonds having flawless  

print("\nPrice Summary:\n", diamonds['price'].describe())
plt.hist(diamonds['price'], bins=50, color='purple', edgecolor='black')
plt.title('Price Distribution (Matplotlib)')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
sns.histplot(diamonds['price'], bins=50, kde=True, color='purple')
plt.title('Price Distribution (Seaborn)')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
# Many diamonds are priced under 5k while some may be larger than 10k

print("\nX Summary:\n", diamonds['x'].describe())
print("\nY Summary:\n", diamonds['y'].describe())
print("\nZ Summary:\n", diamonds['z'].describe())
for dim in ['x', 'y', 'z']:
    plt.hist(diamonds[dim], bins=30, color='teal', edgecolor='black')
    plt.title(f'{dim.upper()} Dimension Distribution (Matplotlib)')
    plt.xlabel(f'{dim.upper()} (mm)')
    plt.ylabel('Frequency')
    plt.show()
    sns.histplot(diamonds[dim], bins=30, kde=True, color='teal')
    plt.title(f'{dim.upper()} Dimension Distribution (Seaborn)')
    plt.xlabel(f'{dim.upper()} (mm)')
    plt.ylabel('Frequency')
    plt.show()
print("\nDepth Summary:\n", diamonds['depth'].describe())
print("\nTable Summary:\n", diamonds['table'].describe())
# Most diamonds have dimensions between 4-9mm with some outliers.


#part 3
#q1 Filter the dataset with carat values between 0.30 & 1.08 divided by the total amount of diamonds
total_diamonds = len(diamonds)
in_range = diamonds[(diamonds['carat'] >= 0.30) & (diamonds['carat'] <= 1.08)]
proportion = len(in_range) / total_diamonds
#0.7468

#q2 Compare both x & y to find where both values are equal
print(f"Proportion between 0.30 and 1.08 carats: {proportion:.4f}")
equal_xy = diamonds[diamonds['x'] == diamonds['y']]
print(f"Diamonds with equal x and y: {len(equal_xy)}") 
#17

#q3 Get the mean of the carats, then filter for all diamonds with a carat below the mean
mean_carat = diamonds['carat'].mean()
below_mean = diamonds[diamonds['carat'] < mean_carat]
print(f"Diamonds with carat less than mean ({mean_carat:.2f}): {len(below_mean)}")
#31127

#q4 Given the cuts a numerical number based on quality, then filter out those that match premium or ideal numerical number
cut_order = {'Fair': 0, 'Good': 1, 'Very Good': 2, 'Premium': 3, 'Ideal': 4}
diamonds['cut_rank'] = diamonds['cut'].map(cut_order)
premium_or_better = diamonds[diamonds['cut_rank'] >= cut_order['Premium']]
print(f"Diamonds with Premium or better cut: {len(premium_or_better)}")
#35342

#q5 Create a column for price per carat and use idmax to find the highest value
diamonds['price_per_carat'] = diamonds['price'] / diamonds['carat']
max_ppc = diamonds.loc[diamonds['price_per_carat'].idxmax()]
print("Diamond with highest price per carat:")
print(max_ppc[['carat', 'price', 'price_per_carat', 'cut', 'color', 'clarity']])
#Diamond with highest price per carat: carat 1.04, price:18542, price per:17828.84


#q6 Want to show price distributions across the cut qualities so we boxplot to visually compare
plt.figure(figsize=(10, 6))
diamonds.boxplot(column='price', by='cut')
plt.title('Diamond Price by Cut (Matplotlib)')
plt.suptitle('')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
plt.figure(figsize=(10, 6))
sns.boxplot(data=diamonds, x='cut', y='price')
plt.title('Diamond Price by Cut (Seaborn)')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
# Fair cuts have more outliers and prices, Premium generally has the higher prices, while ideal has a wide range

#q7 Scatterplot to show the correlation between carat size and price in a visual setting
plt.figure(figsize=(10, 6))
plt.scatter(diamonds['carat'], diamonds['price'], alpha=0.3, s=5)
plt.title('Price vs. Carat (Matplotlib)')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
plt.figure(figsize=(10, 6))
sns.scatterplot(data=diamonds, x='carat', y='price', alpha=0.3, s=10)
plt.title('Price vs. Carat (Seaborn)')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
# Based on the chart the lower the carat, the lower the price, as you go higher in carat so does the price 
#With some outliers among the smaller carats from cut and clarity



#Conclusion
# The dataset of around 54000 diamonds and there attributes, show that a majority of the diamonds are between the carat range .30 & 1.08 with outliers
# Around 38000 are premium or ideal cut, with these higher cuts having higher prices and higher price ranges
# While there is a strong correlation between the carats and the price there are outliers where the larger ones cost disproportionately more.
# Some ouliers exist with the small carats as well which could be attributed to cut and clarity



# References
# 1. Hadley Wickham et al. (2023). ggplot2: Elegant Graphics for Data Analysis.
#    Retrieved from: https://ggplot2.tidyverse.org/reference/diamonds.html
# 
# 2. Pandas Development Team. (2023). Pandas: Python Data Analysis Library.
#    Website: https://pandas.pydata.org/
# 
# 3. Hunter, J. D. (2007). Matplotlib: A 2D Graphics Environment.
#    Computing in Science & Engineering, 9(3), 9095.
# 
# 4. Waskom, M. L. (2021). Seaborn: Statistical Data Visualization.
#    Journal of Open Source Software, 6(60), 3021.

```
